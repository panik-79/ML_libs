# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18, 45, (100))
py_students_age = np.random.randint(18, 40, (100))
```

```python
binss  = [20,25, 30, 35, 40, 45]
plt.hist(ml_students_age, bins = 15 , # or it can take a list also'''
        color = "red", rwidth = 0.8, histtype = "bar", # can also be step
        align = "mid" , orientation = "vertical", #or horizontal
        label = "stud" )
        
plt.title("ML students data")
plt.xlabel("Age category")
plt.ylabel("Number of students")
plt.legend()
plt.show()
```

```python

plt.hist(ml_students_age, # bins can take no. or it can take a list also
        color = "red", rwidth = 0.8, histtype = "bar", # can also be step
        align = "mid" , orientation = "vertical", #or horizontal
        label = "ml_stud" )
plt.hist(py_students_age,  # or it can take a list also'''
        color = "blue", rwidth = 0.8, histtype = "bar", # can also be step
        align = "mid" , orientation = "vertical", #or horizontal
        label = "py_stud" )     
plt.title("Students data")
plt.xlabel("Age category")
plt.ylabel("Number of students")
plt.legend()
plt.show()
```

```python
plt.figure( figsize = (15, 15))
plt.hist([ml_students_age,py_students_age], # bins can take no. or it can take a list also
        color = ["red", "blue"], rwidth = 0.8, histtype = "bar", # can also be step
        align = "mid" , orientation = "vertical", #or horizontal
        label = ["ml_stud", "py_stud"] )
# plt.hist(py_students_age,  # or it can take a list also'''
#         color = "blue", rwidth = 0.8, histtype = "bar", # can also be step
#         align = "mid" , orientation = "vertical", #or horizontal
#         label = "py_stud" )     
plt.title("Students data")
plt.xlabel("Age category")
plt.ylabel("Number of students")
plt.legend()
plt.show()
```
